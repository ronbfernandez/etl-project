```python
# Dependencies
import pandas as pd
```

```python
# Name of CSV file
movie1 = "Resources/imdb_movies.csv"
```

```python
# Import CSV file
movie1_df = pd.read_csv(movie1)
```

```python
# Preview of DataFrame
movie1_df.head()
```

```python
# Deleting extraneous columns
movie1_df.drop(["imdb_title_id", 
                "original_title", 
                "duration", 
                "language", 
                "writer", 
                "date_published", 
                "production_company", 
                "avg_vote", 
                "votes", 
                "metascore", 
                "reviews_from_users", 
                "reviews_from_critics"], axis =1, inplace = True)
movie1_df.head()
```

```python
# Deleting movies from years out of project scope
movie1_df.drop(movie1_df.loc[movie1_df['year']<1988].index, inplace = True)
movie1_df.head()
```

```python
# Identifying incomplete rows
movie1_df.count()
```

```python
# Identifying count on movies with us gross income
movie1_df['usa_gross_income'].value_counts()
```

```python
#removing movies with null/empty us gross income and making new dataset for updates
movies1_final_df = movie1_df[pd.notnull(movie1_df['usa_gross_income'])]
```
